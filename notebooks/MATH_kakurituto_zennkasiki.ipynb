 # 確率と漸化式

参考：[弱点克服　大学生の確率・統計](https://amzn.asia/d/hrIfr0i)

■隣接3項間の漸化式：

$$
a_{n+2}+Aa_{n+1}+Ba_n=0, a_0=C, a_1=D
$$
この解き方は，特性方程式$t^2+At+B=0$の2解を$\alpha, \beta$とすると，
$$
a_n=K\alpha ^n+L\beta ^n
$$

問題１．表が出る確率が$\frac{2}{3}$の効果をn回投げる．表が偶数枚出る確率$p_n$．

問題２．正四面体ABCDの頂点Aに時刻0で粒子がいる．1秒ごとに確率$\frac{1}{6}$ずつで隣の頂点に移動するか，または確率$\frac{1}{2}$で同じ頂点にいるという移動を繰り返す．時刻nで粒子がAにいる確率$p_n$．

問題３．時刻0で粒子が数直線上の原点(0)にいる．硬貨を投げ表が出たら1進み，裏が出たら2進む．$p_n=$数直線上のnを踏む確率．

問題４．表が出る確率が$\frac{2}{3}$である硬貨をn回投げるとき，表が2回続くことはない確率$p_n$．

(1) $p_n$の事象はn回中，表が偶数枚出る確率，$p_{n-1}$の事象はn-1回中，表が偶数枚出る確率である．
よって，求めたい式は，n-1回目で表が出た場合は次に裏を出せばよく，n-1回目で裏が出た場合は次に表を出せばよい．
$$
p_n=p_{n-1}\cdot \frac{1}{3}+(1-p_{n-1})\cdot \frac{2}{3}
$$
$$
x=x\cdot \frac{1}{3}+(1-x)\cdot \frac{2}{3}
$$
これを解くと，$x=\frac{1}{2}$
$p_1=\frac{1}{3}$を用いて，以下の式を解くと，
$$
p_{n+1}-\frac{1}{2}=(-\frac{1}{3})(p_n-\frac{1}{2})
$$
$$
p_n=(-\frac{1}{6})(-\frac{1}{3})^{n-1}+\frac{1}{2}
$$
$$
p_n=\frac{1}{2}(-\frac{1}{3})^n+\frac{1}{2}
$$

(2) $p_n$の事象は時刻nで粒子がAにいる確率，$p_{n-1}$の事象は時刻n-1で粒子がAにいる確率である．
よって，求めたい式は，時刻n-1秒でAにいる場合は確率$\frac{1}{2}$で同じ頂点に，時刻n-1秒でBorCorDにいる場合は確率$\frac{1}{6}$でAに移動すればよい．
$$
p_n=p_{n-1}\frac{1}{2}+3\cdot \frac{(1-p_{n-1})}{3}\cdot \frac{1}{6}
$$
$p_0=1$を用いて上記の式を解くと回答は以下のようになる．
$$
p_n=\frac{1}{4}+\frac{3}{4}(\frac{1}{3})^n
$$

(3) $p_n$の事象はある場所nを踏む確率，$p_{n-1}$の事象はある場所n-1を踏む確率，$p_{n-2}$の事象はある場所n-1を踏む確率である．
よって，求めたい式は，ある場所n-1にいる場合は表を出して1進めばよく，ある場所n-2にいる場合は裏を出して2進めばよい．
$$
p_n=\frac{1}{2}p_{n-1}+\frac{1}{2}p_{n-2}
$$
$p_1=\frac{1}{2}, p_2=\frac{1}{2}+(\frac{1}{2})^2=\frac{3}{4}$であるので，解説の特性方程式を用いて式を解くと，以下のようになる．
$$
p_n=\frac{2}{3}+\frac{1}{3}(-\frac{1}{2})^n
$$

(4) $q_n$の事象を初めに表が出るという現象の元で，n回硬貨を投げた時に表が2回続かない確率とする，また，$p_n$の事象を表裏どちらが出てもよいが表が2回続かない確率とする．
$$
q_n=\frac{2}{3}\cdot 0+p_{n-1}\frac{1}{3}
$$
$$
p_n=\frac{2}{3}q_{n-1}+\frac{1}{3}p_{n-1}
$$
よって，
$$
p_n=\frac{2}{3}p_{n-2}+\frac{1}{3}p_{n-1}，p_1=1, p_2=1-(\frac{2}{3})^2=\frac{5}{9}
$$
これを解くと解答は以下のようになる．
$$ 
p_n=\frac{4}{3}(\frac{2}{3})^n+(-\frac{1}{3})^{n+1}
$$
