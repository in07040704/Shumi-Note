# 同時確率分布と共分散

参考：[弱点克服　大学生の確率・統計](https://amzn.asia/d/hrIfr0i)

■$(X,Y)$の同時確率分布：
$$
1=\sum_{i,j}P(X=i \cap Y=i)
$$
上の全確率(標本空間全てに対する確率)が成立する．また全ての2変数関数$h(x,y)$に対して，
$$
E(h(X,Y))=\sum_{i,j}h(i,j)P(X=i\cap Y=j)
$$
で計算する．同時確率分布$P(X=i\cap Y=j)$が与えられれば個々の確率分布$P(X=i),P(X=j)$は，
$$
P(X=i)=\sum_{j}P(X=i\cap Y=j),  P(Y=j)=\sum_{i}P(X=i\cap Y=j)
$$

■$(X,Y)$の共分散：

$m=E(X), m'=E(Y)$とおいたとき，
$$
Cov(X,Y)=E((X-m)(Y-m'))=E(XY)-mm'-mm'+mm'=E(XY)-E(X)E(Y)
$$
を$(X,Y)$の**共分散**と呼ぶ．

$$
\rho(X,Y)=\frac{Cov(X,Y)}{\sigma(X)\sigma(Y)}=Cov(Xの標準化，Yの標準化)
$$
を$(X,Y)$の**相関係数**と呼ぶ．
