# 等差数列・等比数列・シグマ計算

参考：[弱点克服　大学生の確率・統計](https://amzn.asia/d/hrIfr0i)

■等差数列：


等差数列とは，$a_{n+1}-a_n=d$ （公差が一定）満たされるときである．等差数列の和の公式は，
$$
\sum_{初}^{末}等差数列 = \frac{初項+末項}{2} \times 項数
$$

■等比数列：
等比数列とは，$b_n=rb_{n-1}$ (rは公比が一定)のとき，$b_n=b_1r^{n-1}=b_0r^n$ である．等比数列の和の公式は，
$$
\sum_{初}^{末}等比数列 = \frac{初項-末項\times 公比}{1-公比}
$$

また，無限等比級数は
$$
\sum_{初}^{\infty}等比数列 = \frac{初項}{1-公比} (|公比|<1 のとき)
$$

■シグマ($\sum$)計算：
$$
\sum_{k=1}^{n}c=cn
$$

$$
\sum_{k=1}^{n}c=c(n+1)
$$

$$
\sum_{k=1}^{n}k=\frac {n(n+1)}{2}
$$

$$
\sum_{k=1}^{n}k^2=\frac {n(n+1)(2n+1)}{6}
$$

$$
\sum_{k=1}^{n}k^3=(\frac {n(n+1)}{2})^2
$$

問題１．$a_0=1$として$a_{n+1}=3a_{n+1}+4$を解け．

問題２．$a_0=1$として$a_{n+1}=3a_{n+1}+n$を解け．

(1) 定数$C$を特殊解とすると，$C=3C+4$となる．$C=-2$であることから，$a_{n+1}-C=3(a_{n}-C)$と$a_0-C=1+2=3$から，
$$
a_n+2=3\cdot3^n つまり，a_n=3^{n+1}-2
$$

(2) 特殊解$b_n$ として，$b_n=an+b$とおくと，$a(n+1)+b=3(an+b)+n$と置き換わる，両辺でnの次数と項を比較すると，$a=-\frac{1}{2}, b=-\frac{1}{4}$であることが分かるので，$b_n=-\frac{n}{2}-\frac{1}{4}$.与えられた式より，$a_{n+1}-b_{n+1}=3(a_n-b_n)$なので，$a_n-b_n=(a_0-b_0)3^n$ つまり，
$$
a_n=\frac{5}{4}3^n-\frac{1}{2}n-\frac{1}{4}
$$
