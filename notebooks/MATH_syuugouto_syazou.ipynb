# 集合と写像

参考：[弱点克服　大学生の確率・統計](https://amzn.asia/d/hrIfr0i)

■１対１写像(単射)：


$ f(x)=f(y) $なら $ x=y $ である写像を **1対1の写像(単射)** と呼ぶ．

■上への写像(全射)：


任意の $ y \in B $に対してある $ x \in A $ が存在して $y=f(x)$ となるとき，写像$f$は **上の写像(全射)** と呼ぶ．

■逆写像(逆関数)：


任意の $ x \in B $に対して　$y=f(x)$ となる $ y \in A $ が一意的に存在する．その $y$ への対応を $y=f(x)^{-1}$ と書き，この　$f(x)^{-1}$ を $f$ の**逆写像(関数の場合は逆関数)** と呼ぶ．

問題１．$\{ {a,b,c}\}$ を値域，$\{{A,B,C,D,E}\}$ を定義域とする写像は何個あるか？また，上への写像は何個あるか？

問題２．$\{{1,2,3,4,5}\}$ から $\{{1,2,3,4,5,6,7,8}\}$ への写像で全ての $i < j (i,j \in \{{1,2,3,4,5}\})$ に対し $f(i) \leqq f(j)$ を満たす写像の個数を求めよ．

(1) 写像の組み合わせは $3^5$ である．$ \# (Im(f)) =1 $ である写像の個数は3個．$ \# (Im(f)) =2 $ である写像の個数は $3 \times (2^5 -2)$ である．残りは上への写像となるので，求める上への写像の個数は
$$
 = 3^5-(2^5-2)-3 = 150
$$ 
である．

(2) 5個を8種類の写像に分配するので，重複組み合わせである．
$$
_8H_5 =792
$$
