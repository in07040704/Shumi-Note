 # じゃんけんの問題

参考：[弱点克服　大学生の確率・統計](https://amzn.asia/d/hrIfr0i)

■多項展開の式：

$$
(a+b+c)^n=\sum_{k,l\geqq 0, k+l\leqq n}\frac{n!}{k!l!(n-(k+l))!}a^kb^lc^{n-k-l}
$$

問題１．2人でじゃんけんをするとき，あいこになる確率．

問題２．3人でじゃんけんをするとき，あいこになる確率．

問題３．6人でじゃんけんをするとき，グー・チョキ・パーが2人ずつになる確率$p$，勝者が1人出る確率$q$，勝者が2人出る確率r．

問題４．3人でじゃんけんをするとき，負けたら次のじゃんけんに参加できないとしてn回目のじゃんけんで初めて1人の勝者が出る確率$p_n$．

(1) 2人でじゃんけんをしてあいこになるのは，2人とも，グー・チョキ・パーを出す時であるので，全事象の合計は9通りから，解答は$\frac{3}{9}=\frac{1}{3}$．

(2) 3人でじゃんけんをしてあいこになるのは，3人とも，グー・チョキ・パーを出す時と3人がそれぞれ違う手を出す6通りがあるので，合計は9通りである．全事象の合計は27通りから，解答は$\frac{9}{27}=\frac{1}{3}$．

(3) 6人でじゃいけんをして，グー・チョキ・パーをそれぞれ2人が出す確率なので，手の出し方を6人に割り振ると$\frac{6!}{2!2!2!}$通りとなる．よって求める確率は，以下のように計算される．$\frac{6!}{2!2!2!}(\frac{1}{3})^6$

次に，勝者が1人出る確率$q$と勝者が2人出る確率$r$は，たとえば，グーで1人の勝者が出る確率を考えて3倍すればよい．
$q=3P(グー=1\cap チョキ=5)=3\frac{6!}{5!}(\frac{1}{3})^6=\frac{2}{81}$

$r=3P(グー=2\cap チョキ=4)=3\frac{6!}{2!4!}(\frac{1}{3})^6=\frac{5}{81}$

(4) $k(1\leqq k\leqq n-1)$回目のじゃんけんで3人が2人になったとし，その後n回目で初めて勝者が出る確率は，まず，3人でじゃんけんをして1人抜ける全事象はあいこが全員がグー・チョキ・パーがかぶる３通りと全員が異なる手を出す６通りあり，２人抜けるのは$3P(グー=2\cap チョキ=1)=9$通り，１人抜けるのは$3P(グー=1\cap チョキ=2)=9$通りある．よって，3人から1人が抜ける確立は$\frac{9}{9+9+9}=\frac{1}{3}$.次に，2人でじゃんけんをして1人抜ける全事象はあいこが2人ともグー・チョキ・パーがかぶる３通り，どちらかが勝つのは$3P(グー=1\cap チョキ=1)=6$通りある．よって，2人から1人が抜ける確立は$\frac{6}{6+3}=\frac{2}{3}$.
以上から，その後n回目で初めて勝者が出る確率は，$(\frac{1}{3})^{k-1}(\frac{1}{3})(\frac{1}{3})^{n-k-1}(\frac{2}{3})=(\frac{1}{3})^{n-1}(\frac{2}{3})$

よって，以上の確率は$1～n-1$回目の全てで成り立つので，全事象はn-1まで全てを足せばよいので，$(n-1)(\frac{1}{3})^{n-1}(\frac{2}{3})$である．また，n-1回目まで3人でn回目に初めて1人の勝者が出る確率は$(\frac{1}{3})^{n-1}(\frac{1}{3})=(\frac{1}{3})^n$足し合わせて，求める確率は，
$$
(2n-1)(\frac{1}{3})^n
$$
