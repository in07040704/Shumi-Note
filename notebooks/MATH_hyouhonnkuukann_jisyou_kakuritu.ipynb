 # 標本空間・事象・確率

参考：[弱点克服　大学生の確率・統計](https://amzn.asia/d/hrIfr0i)

■全事象の確立：

$$
P(\Omega)=1
$$

■加法性：

$$
A\cap B=0(AとBが排反)のとき，
P(A\cup B)=P(A)+P(B)
$$
これが満たされれば，**根元事象(有限標本空間**の場合は1点からなる事象$\{1\},\{2\},\{3\},\{4\},\{5\},\{6\}$のそれぞれに確率が与えられれば，他の全ての事象の確率を計算することが出来る．
$$
P(\{1\})=P(\{2\})=\cdot \cdot \cdot =P(\{6\})=\frac{1}{6}
$$
と与えられれば，他の事象(例えば，$P(\{2,3,4\})=P(\{2\})+P(\{4\})+P(\{6\})=\frac{3}{6}=\frac{1}{2}$のように)の確率も決まる．

また，事象$P(\{2,3,4\})\Leftrightarrow \times \bigcirc\times \bigcirc\times \bigcirc$を対応させれば，$\times ,\bigcirc$６つの列と事象は1対1対応する(事象$\leftrightarrow 「\{1,2,3,4,5,6\}から\{\times ,\bigcirc\}$ への写像」が1対1といってもよい)．つまり，$\#(\mathcal{F})=事象の個数=2^6(=2^{\#(\Omega)})$である．

問題１．サイコロを１個投げる確率空間(標本空間 $(\Omega)$・全事象$(\mathcal{F})$・確率$(P)$)を書け．

問題２．1の目の出る確率＝$\frac{1}{2}$で他は均等であるインチキなサイコロを１個投げる確率空間を書け．

問題３．サイコロを2個(2回)投げる確率空間を書け．

問題４．表が出る確率が$p$であるインチキな硬貨を3回投げる確率空間を書け．

問題５．1から$N$までの整数が1つずつ書かれた$N$枚のカードを1枚ずつ元に戻さないで3枚選ぶ．確率空間(ただし，$N\leqq 3$)を書け．

問題６．問題1~5で，$\#(\Omega)=\Omega$の要素の個数とする時，$\#(\mathcal{F})$を求めよ．

(1) $$\Omega=\{1,2,3,4,5,6\}$$
$$\mathcal{F}=\{0,\{1\},\{2\},\{3\},\{4\},\{5\},\{6\},\{1,2\},\{1,3\},\{1,4\},\{1,5\},\{1,6\},\{2,3\},\{2,4\},\{2,5\},\{2,6\},\{3,4\},\{3,5\},\{3,6\},\{4,5\},\{4,6\},\{5,6\},\cdot \cdot \cdot \{2,3,4,5,6\},\Omega\}$$
$$P(\{1\})=P(\{2\})=\cdot \cdot \cdot =P(\{6\})=\frac{1}{6}$$

(2) $\Omega,\mathcal{F}$は，(1)と同じ．
$$P(\{1\})=\frac{1}{2}$$
$$P(\{2\})=\cdot \cdot \cdot =P(\{6\})=\frac{1}{10}$$

(3) $\Omega=\{(i,j)|1\leqq i\leqq 6,1\leqq j\leqq 6,i,j\in \mathbb{N}\}$

$\mathcal{F}=\Omega$の部分集合全体

$P(\{(i,j)\})=\frac{1}{36},1\leqq i \leqq 6, 1\leqq j \leqq 6,i,j\in \mathbb{N}$

(4) $P(\{表,表,表\})=p^3$, 

$P(\{表,表,裏\})=P(\{表,裏,表\})=P(\{裏,表,表\})=p^2(1-p)$, 

$P(\{表,裏,裏\})=P(\{裏,裏,表\})=P(\{裏,表,裏\})=p(1-p)^2$

$P(\{裏,裏,裏\})=(1-p)^3$

(5) $\Omega=\{(i,j,k)|1\leqq i\leqq N,1\leqq j\leqq N,1\leqq k\leqq N, i\neq j, j\neq k, k\neq i\}$

$\mathcal{F}=\Omega$の部分集合全体
$P(\{(i,j,k)\})=\frac{1}{N(N-1)(N-2)}$

$1\leqq i,j,k\leqq N, i\neq j, j\neq k, k\neq i$

(6) (1),(2)については，
$$\#(\Omega)=6, \#(\mathcal{F})=2^6=64$$
(3)については，
$$\#(\Omega)=36, \#(\mathcal{F})=2^36$$
(4)については，
$$\#(\Omega)=8, \#(\mathcal{F})=2^8=256$$
(5)については，
$$\#(\Omega)=N(N-1)(N-2), \#(\mathcal{F})=2^{N(N-1)(N-2)}$$
